```python
max_num = 26 # a-z
graph = [[''] * max_num] * max_num
adj = [[0] * max_num] * max_num
indegree = outdegree = [0] * max_num
```

```python
def makeGraph(words):    
    for i in range(len(words)):
        a = ord(words[i][0]) - ord('a')
        b = ord(words[i][len(words[i])-1]) - ord('a')
        graph[a][b] = words[i]
        adj[a][b] += 1
        outdegree[a] += 1
        indegree[b] += 1
```

```python
def getEulerCircuit(u, circuit):
    for v in range(len(adj)):
        while adj[u][v] > 0:
            adj[u][v] -= 1
            getEulerCircuit(v, circuit)
    circuit.append(u)
```

```python
def getEculterTrailOrCircuit():
    circuit = []
    for i in range(max_num):
        if outdegree[i] == indegree[i] + 1:
            getEulerCircuit(i, circuit)
            return circuit
#     for i in range(max_num):
#         if outdegree[i]:
#             getEulerCircuit(i, circuit)
#             return circuit
    return circuit
```

```python
def checkEuler():
    plus1 = 0
    minus1 = 0
    for i in range(max_num):
        delta = outdegree[i] - indegree[i]
        if delta < - 1 or delta > 1:
            return False
        if delta == 1:
            plus1 += 1
        if delta == -1:
            minus1 += 1
    return (plus1 == 1 and minus1 == 1) or (plus1 == 0 and minus1 == 0)
```

```python
def solve(words):
    makeGraph(words)
    if checkEuler() == False:
        return "IMPOSSIBLE"
    circuit = getEculterTrailOrCircuit()
    if len(circuit) != len(words) +1:
        return "IMPOSSIBLE"
    circuit = circuit[::-1]
    
    result = ''
    for i in range(1, len(circuit)):
        a = circuit[i - 1]
        b = circuit[i]
        if len(result):
            result += " "
        result += graph[a].pop(b)
    return result
```

```python
solve(words)
```
